# SQD - Sample based Quantum Diagonalization

Sample based Quantum Diagonalization - è un algoritmo che serve a trovare autovalori e autovettori di matrici Hamiltoniane molto grandi, e che trova applicazioni nella chimica quantistica: permette infatti di trovare, ad esempio, l'energia dello stato fondamentale di molecole o modelli a più corpi.

## Approccio in breve

1. **Definizione del problema** - si mappa il proprio Hamiltoniano su dei qubit tramite opportune trasformazioni. In chimica, si può usare la **trasformazione di Jordan-Wigner**, che mappa molecole a operatori quantistici (Hamiltoniana molecolare -> Operatore fermionico -> Operatore).
2. **Creazione degli stati da testare** - si usa un ansatz (una *guess* in funzione del problema) sotto forma di circuito per inizializzare gli stati del circuito.
3. **Esecuzione e misurazione dei risultati**.
4. **Costruzione del sottospazio** in funzione dei valori misurati.
5. **Diagonalizzazione classica** - ma su una matrice di dimensione molto più piccola!
