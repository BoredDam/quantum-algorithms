# Entangling and GHZ states

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram

from BoredQiskit import FastQiskit as fq
```

## Simple Entanglement

Entanglement simmetrico $|\Phi^+\rangle$.

```python
qc = QuantumCircuit(2,2)
qc.h(0)
qc.cx(0,1)
qc.measure(range(2), range(2))
qc.draw('mpl')
```

```python
counts = fq.run(qc)
plot_histogram(counts)
```

## Circuiti GHZ

```python
qc = QuantumCircuit(3,3)
qc.h(0)
qc.cx(0,1)
qc.cx(0,2)
qc.measure(range(3), range(3))
qc.draw('mpl')
```

```python
counts = fq.run(qc)
plot_histogram(counts)
```

### GHZ su simulatore realistico

```python
counts = fq.torino_run(qc)
plot_histogram(counts)
```

## Generalized GHZ

```python
def generalized_ghz(q) -> QuantumCircuit:
    if q < 3:
        q = 3
    if q > 30:
        q = 30
    qc = QuantumCircuit(q, q)
    qc.h(0)
    qc.cx(range(q-1),range(1,q))
    qc.barrier()
    qc.measure(range(q), range(q))
    return qc
```

```python
qc = generalized_ghz(6)
qc.draw('mpl')
```

```python
fq.run(qc)
```
