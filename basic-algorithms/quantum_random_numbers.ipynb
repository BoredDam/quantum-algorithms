# Some basic random number generators

```python
from qiskit import QuantumCircuit
from BoredQiskit import FastQiskit as fq
```

## Quantum Coin

```python
qc = QuantumCircuit(1,1)
qc.h(0)
qc.measure(0,0)
qc.draw('mpl')
```

```python
fq.run(qc)
```

## Quantum Random Number

```python
def rand_number_quantum_circuit(n=1):
    qc = QuantumCircuit(n,n)
    qc.h(range(n))
    qc.measure(range(n), range(n))
    return qc
```

```python
qc = rand_number_quantum_circuit(2)
qc.draw('mpl')
```

```python
result = fq.run(qc, 1)
result = list(list(result)[0])

num = 0
pos = 0
for i in result:
    num += int(i) * 2 ** pos
    pos += 1

print(num)
```
